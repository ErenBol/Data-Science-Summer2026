```python
# Creating a df
people = {
    "first": ["Corey", "Jane", "John", "Mary"],
    "last": ["Schafer", "Doe", "Doe", "Smith"],
    "email": ["Corey.Schafer@gmail.com", "Jane.Doe@gmail.com", "John.Doe@gmail.com", "Mary.Smith@gmail.com"]
}

print(people["last"])

people_df = pd.DataFrame(people)
people_df
```

```python
print(people_df["email"]) ## or df.email; prefer 1st
print(type(people_df["email"]))  ## Columns are Series objects !!!!
```

```python
people_df[['first', 'last']]
type(people_df[['first', 'last']])  ## DataFrame object !!!
```

```python
## Getting a row : loc and iloc

#iloc is used to get rows (and columns) at particular positions in the index (so it only takes integers)
print(people_df.iloc[0])  ## 0th row
print(people_df.iloc[[0, 1]])  ## 0th and 1st rows
print(people_df.iloc[[0, 1],  [0,2]])  ## 0th and 1st rows, 0th and 2nd columns

#loc is used to get rows (and columns) with particular labels from the index. 
# It can also be used with a boolean array and with labels of the columns.
# End is inclusive for loc, but exclusive for iloc
people_df.loc[0]  ## 0th row
people_df.loc[[0, 1], ["first"]]  ## 0th and 1st rows
print(people_df.loc[people_df["last"] == "Doe"])  ## all rows where last name is Doe
people_df.loc[people_df["last"] == "Doe", ["first", "email"]]  ## all rows where last name is Doe, and only first and email columns
```
