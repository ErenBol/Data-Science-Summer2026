```python
## Missing values
df.isna()              # True/False missing map
print(df.isna().sum())        # missing count per column
df.dropna()            # remove rows with missing values, needs assignment
df.fillna(0)           # replace missing values with 0, needs assignment
print(df["WorkExp"].isna().sum())  # missing count for WorkExp column
df["WorkExp"] = df["WorkExp"].fillna(0)  # replace missing values in WorkExp column with 0, change the column itself (inplace doesnt work for single column)
print(df["WorkExp"].isna().sum())  # missing count for WorkExp column after filling missing values with 0

```

```python
## Adding/modifying columns
df["NewCol"] = 10  # add a new column, all rows will have value 10 in this column
df["NewCol2"] = df["WorkExp"] * 2  # add a new column, values are double of WorkExp column
df["NewCol3"] = df["NewCol"] + df["NewCol2"]  # add a new column, values are sum of NewCol and NewCol2 columns
df["NewCol4"] = np.where(df["WorkExp"] >25, "Experienced", "Less Experienced")  # add a new column, values are "Experienced" if WorkExp > 25, otherwise "Less Experienced"
print(df[["WorkExp", "NewCol", "NewCol2", "NewCol3", "NewCol4"]].head())
```

```python
## apply, map, applymap (deprecated, don't use), replace,  lambda functions
print(people_df)

# apply: apply a function along an axis of the DataFrame (rows or columns)
print(people_df["email"].apply(len)) # apply a function to each element of the Series (column)

def update_email(email):
    return email.upper()
print(people_df["email"].apply(update_email))

# lambda: anonymous function
print(people_df["email"].apply(lambda x: x.lower()))

#applying to df itself
print(people_df.apply(len, axis= "columns")) # return number of columns for each row
print(people_df.apply(pd.Series.min, axis=0))  #default axis=0 # return minimum value for each column

# replace: replace values in a Series or DataFrame
print(people_df["last"].replace({"Doe": "Smith", "Smith": "Johnson"}))  # using map results in NaN for values not in the map
```

```python
## String operations
print(people_df)
print(people_df["first"].str.lower())
print(people_df["first"].str.upper())
print(people_df["first"].str.strip())
print(people_df["first"].str.contains("john", case=False, na=False)) # Check if each value contains "abc" case insensitive, and treat NaN as False
print(people_df["first"].str.replace("John", "newJohn")) # Can use dictionary to replace multiple values, or regex for pattern matching
print(people_df["first"].str.split(" "))
```

```python
## Adding rows: using loc and concat (append is deprecated, don't use)

# Add one row
people_df.loc[len(people_df)] = ["John", "Doe", "John.Doe@gmail.com"]
print(people_df)

# Add one row with dictionary
people_df.loc[len(people_df)] = {
    "first": "Mary",
    "last": "Smith",
    "email": "Mary.Smith@gmail.com"
}
print(people_df)

# Add multiple rows (concat a DataFrame)
new_rows = pd.DataFrame([
    {"first": "Ali", "last": "Yilmaz", "email": "ali@gmail.com"},
    {"first": "Ayse", "last": "Demir", "email": "ayse@gmail.com"}
])

people_df = pd.concat([people_df, new_rows], ignore_index=True) # always use ignore_index=True to prevent duplicate indexes
print(people_df)
```

```python
### Grouping and aggregation
print(df.shape)  # number of rows in the DataFrame
print(df["Employment"].nunique())  # number of unique values in Employment column
print(df["Employment"].count())  #  count of not null values in Employment column
print(df["Employment"].value_counts())  # count of unique values in Employment column
print(df["Employment"].value_counts(normalize=True))  # percentage of unique values in Employment column

print("*"*100)
print(df.groupby("Employment").head().loc[:, ["Employment", "WorkExp"]].sort_values("Employment"))  # first 5 rows of each group, and only Employment and WorkExp columns, sort for better readability
# Means: 
# 1. group by Employment column
# 2. take the first 5 rows of each group
# 3. select only Employment and WorkExp columns for all rows
# 4. sort the result by Employment column for better readability 
# Cleaner way:
print("*"*100, "Cleaner way:")
result = df.groupby("Employment").head(5)
result = result.sort_values("Employment")

print(result.loc[:, ["Employment", "WorkExp"]])

# Aggregation functions: count, mean, median, std, min, max, sum, etc.
print("*"*100, "Aggregation:")
print(df.groupby("Employment")["WorkExp"].mean())  # mean of WorkExp column for each group in Employment column

print("*"*100)
print(df.groupby("Employment")["WorkExp"].agg(["count", "mean", "median", "std"]))  # count, mean, median, std of WorkExp column for each group in Employment column
## Another usage: different operations for different columns and naming those aggregated columns !!
    # print(df.groupby("Employment").agg(
    #     count_column = ("colum1" , "count"), 
    #     avg_column = ("WorkExp", "mean"),
    #     sum_column = ("colum2" , "sum")
    # ))
```

```python
# Important patterns:
# 1. Aggregate on a Series
df["WorkExp"].mean()
df["WorkExp"].sum()
df["WorkExp"].min()
df["WorkExp"].max()
df["WorkExp"].count()
df["WorkExp"].agg(["count", "mean", "min", "max"])

# 2. Aggregate on a DataFrame
df.mean(numeric_only=True)
df.sum(numeric_only=True)
df.min()
df.max()
df[["WorkExp", "ConvertedCompYearly", "JobSat"]].agg(["count", "mean", "min", "max"])

# 3. Aggregate after groupby
df.groupby("Employment")["WorkExp"].mean()
df.groupby("Employment")["WorkExp"].agg(["count", "mean", "min", "max"])
df.groupby("Employment")[["WorkExp", "JobSat"]].agg(["count", "mean"])
```

```python
# Example: Find the percentage of people that has experience more than 10 years in each employment group
group = df.groupby("Employment")["WorkExp"]
group_aggregated = group.agg(lambda x: (x >10).mean() * 100)
df_aggregated = group_aggregated.reset_index(name = "percentage")
df_aggregated

# Important: Pandas calls the lambda once per group, not once per row.
# So x is all the WorkExp values for that group, not a single value.
# Therefore x > 10 returns a boolean array for all values in that group.
```

```python
## Merging dataframes (SQL join like operations)
people1 = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

people2 = {
    "first": ["Corey", "Jane", "John", "Mary"],
    "last": ["Schafer", "Doe", "Doe", "Smith"],
    "email": ["Corey.Schafer@gmail.com", "Jane.Doe@gmail.com", "John.Doe@gmail.com", "Mary.Smith@gmail.com"],
    "age": ["33", "55", "63", "28"]
}

df1 = pd.DataFrame(people1)
df2 = pd.DataFrame(people2)

print(df1.head())
print(df2.head())

df_merged1 = pd.merge(df1, df2, on="first")
df_merged2 = pd.merge(df1, df2, on="first", how="inner")
df_merged3 = pd.merge(df1, df2, on="first", how="left")

print(df_merged1.head())
print(df_merged2.head())
print(df_merged3.head())

```

```python
# using format strings: 
# %Y: 4-digit year, %m: 2-digit month, %d: 2-digit day, 
# %I: 2-digit hour (12-hour clock),
# %H: 2-digit hour (24-hour clock), %p: AM/PM
stock_df["date"] = pd.to_datetime(stock_df["date"], format="%Y-%m-%d %I-%p")
stock_df["date"]
```
